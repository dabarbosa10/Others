```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datos=np.loadtxt('OpD.dat')
```

```python
datos.shape
```

```python
r=datos[:,1]
T=datos[:,2]
rho=datos[:,3]
kappa=datos[:,4]
```

```python
r[1]
```

```python
def Traptao(ka,rh,r1):
    Tau=np.zeros(len(r1))
    for i in range(len(r1)-1):
        Tau[i+1]=-((ka[i]*rh[i]+ka[i+1]*rh[i+1])/2)*(r1[i+1]-r[i])+Tau[i]
    return Tau
```

```python
solTao=Traptao(kappa, rho,r)
```

```python
plt.figure(figsize=(15,15))
plt.plot(solTao,T,'k*-')
plt.xlabel('Optical Depth' ,fontsize=14)
plt.ylabel('Temperature T(K)',fontsize=14)
plt.grid()
```

```python
def temp(taun):
    Te=5504
    T1=Te*np.power(((3.0/4.0)*(taun+2/3)),0.25)
    return T1
```

```python
temperature=temp(solTao)
```

```python
plt.figure(figsize=(15,15))
plt.plot(solTao,T,'k*-', label='Trapezoidal Method')
plt.plot(solTao,temperature,'r+-',label='Using equation')
plt.xlabel('Optical Depth' ,fontsize=14)
plt.ylabel('Temperature T(K)',fontsize=14)
plt.legend(fontsize=20)
plt.grid()
plt.savefig('Tarea1.pdf')
```
