```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Pfin(Ri,Rf,P1):
    Pf=P1*np.power((Ri/Rf),5)
    return Pf
```

```python
def Rfin(Ri,vf,deltaT):
    Rf=Ri+vf*deltaT
    return Rf
```

```python
def Vfin(vi,Pi,Ri,M,G,m,deltaT):
    Vf=vi+((4*np.pi*np.power(Ri,2))*Pi/m-(G*M)/np.power(Ri,2))
    return Vf
```

```python
def Calculation(time):
    M=1e31
    m=1e26
    R1=1.7e10
    P1=5.6e4
    G=6.67408e-11
    deltate=time
    vin=0
    Vf=np.zeros(151)
    Rf=np.zeros(151)
    Pf=np.zeros(151)
    Vf[0]=0
    Rf[0]=R1
    Pf[0]=P1
    T=np.linspace(0,1.5e7,151)
    for i in range(1,len(T)):
        Vf[i]=Vf[i-1]+ ((4*np.pi*np.power(Rf[i-1],2)*Pf[i-1])/m -(G*M/np.power(Rf[i-1],2)))*(1e4)
        Rf[i]=Rf[i-1]+Vf[i]*(1e4)
        Pf[i]=Pf[i-1]*np.power((Rf[i-1]/Rf[i]),5)
    return(T,Vf,Rf,Pf)  
    
```

```python
Time,Vel,Rad,Pres=Calculation(1e4)
```
