```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.core.pylabtools import figsize
```

```python
figsize(12.5,4)
plt.rcParams['savefig.dpi']=300
plt.rcParams['figure.dpi']=300
```

```python
colors=["#348ABD","#A60628"]
prior=[1/21,20/21]
posterior=[0.087,1-0.087]
plt.bar([0,0.7],prior,alpha=0.70,width=0.25,color=colors[0],label="Prior distribution",lw="3",edgecolor="#348ABD")
plt.bar([0+0.25,0.7+0.25],posterior,alpha=0.70,width=0.25,color=colors[1],label="Posterior distribution",
        lw="3",edgecolor="#A60628")
plt.grid()
plt.xticks([0.20,0.95],["Librarian","Farmer"])
plt.title("Prior and Posterior distributions")
plt.ylabel("Probability")
plt.legend(loc="upper left")
```

```python
figsize(12.5,4)
import scipy.stats as stats
a=np.arange(20)
poi=stats.poisson
lambda_=[1.5,4.5]
colors=["#348ABD","#A60628"]
plt.bar(a,poi.pmf(a,lambda_[0]),color=colors[0],label="$\lambda=%.1f"%lambda_[0],
       alpha=0.60,edgecolor=colors[0],lw="3")
plt.bar(a,poi.pmf(a,lambda_[1]),color=colors[1],label="$\lambda=%.1f"%lambda_[1],
       alpha=0.60,edgecolor=colors[1],lw="3")
plt.grid()
plt.xticks(a+0.4,a)
plt.legend()
plt.ylabel("Probability of k")
plt.title("PMF of a Poisson random variable")
```

```python
figsize(11,9)
dist=stats.beta
n_trials=[0,1,2,3,4,10,100,1000]
data=stats.bernoulli.rvs(0.5,size=n_trials[-1])
x=np.linspace(0,1,100)
```

```python
for k, N in enumerate(n_trials):
    sx = plt.subplot(len(n_trials)/2, 2, k+1)
    plt.xlabel("$p$, probability of heads") \
        if k in [0, len(n_trials)-1] else None
    plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1 + N - heads)
    plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)
    plt.grid()


plt.suptitle("Bayesian updating of posterior probabilities",
             y=1.02,
             fontsize=14)

plt.tight_layout()
```
