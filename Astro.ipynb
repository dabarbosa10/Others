```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0.02,6,1000)
```

```python
D0=1-(np.power(x,2))/6
D1=np.sin(x)/(x)
D5=np.power((1+np.power(x,2)/3),-1/2)
```

```python

plt.figure(figsize=(13,10))
plt.plot(x,D0,label='$ \\theta_0$')
plt.plot(x,D1,label='$ \\theta_1$')
plt.plot(x,D5,label='$ \\theta_5$')
plt.grid()
plt.legend(fontsize=14)
plt.xlim(0,6)
plt.ylim(0,1)
plt.xlabel('$\\xi$' ,fontsize=16)
plt.ylabel('$\\theta_n (\\xi)$',fontsize=16)
plt.savefig('Politropic.png')
```

```python
phi=np.sqrt(6-3*(np.power(x,4)-2*np.power(x,2)-2)*np.exp(-np.power(x,2)))
```

```python
plt.plot(x,phi)
```

```python
roc=9*10**4
```

```python
Pc=1.9*10**16
Tc=16*10**6
```

```python
Ro_C=roc*(np.power(x,3)*np.exp(-np.power(x,2)))/phi
```

```python
P=Pc*np.exp(-np.power(x,2))
T=Tc*(phi/np.power(x,3))
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,Ro_C,label='Densidad')
plt.xlabel('$x=r/a$' ,fontsize=16)
plt.ylabel('$\\rho (x)$',fontsize=16)
plt.grid()
plt.xlim(0,6)
plt.legend()
plt.savefig('Density.png')
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,P,'r',label='Presion')
plt.xlabel('$x=r/a$' ,fontsize=16)
plt.ylabel('$P (x)$',fontsize=16)
plt.grid()
plt.xlim(0,6)
plt.legend()
plt.savefig('Pressure.png')
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,T,'k',label='Temperatura')
plt.xlabel('$x=r/a$' ,fontsize=16)
plt.ylabel('$T (x)$',fontsize=16)
plt.grid()
plt.xlim(0,6)
plt.legend()
plt.savefig('Temperature.png')
```
